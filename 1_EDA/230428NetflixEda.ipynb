Number of genres in Netflix

```python
df_movie1['listed_in'] = df_movie1['listed_in'].astype('string')
```

```python
df_movie1['listed_in_set'] = np.nan
for index,row in df_movie1.iterrows():
    df_movie1['listed_in_set'][index] = row['listed_in'].split(', ')
```

```python
df_movie1.head()
```

```python
set_of_genres = set()
for index,row in df_movie1.iterrows():
    set_of_genres.update(row['listed_in_set'])
set_of_genres
```

```python
genres = []
genre_counts=[]
for genre in set_of_genres:
    count = 0
    for index,row in df_movie1.iterrows():
        if genre in row['listed_in_set']:
            count += 1
    print(f'The number of {genre} genre movies is {count}')
    genres.append(genre)
    genre_counts.append(count)
```

```python
genre_series = pd.Series(genre_counts, index=genres)
genre_series.sort_values()
```

Country wise movies count

```python
country_list_combined = []
for index,row in df_movie1.iterrows():
    country_list_combined.extend(row['country_list'])
country_list_combined
```

```python
country_set = set(country_list_combined)
country_set
```

```python
len(country_set)
```

```python
country_count_data = []
for country in country_set:
    temp = country_list_combined.count(country)
    country_count_data.append(temp)
country_list = list(country_set)
```

```python
count_series = pd.Series(country_count_data, index=country_list)
max(count_series)
```

```python
len(count_series)
```
