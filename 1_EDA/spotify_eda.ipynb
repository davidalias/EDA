# Data Analysis

Top 5 most popular artists

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
top_artists= df.groupby('artist').count().sort_values(by='song_title',ascending=False)['song_title'][:5]
top_artists
```

```python
top_artists.plot.barh()
plt.show()
```

Top 5 loudest tracks

```python
loud_tracks = df.sort_values(by='loudness', ascending = True)[['song_title','loudness','artist']][:5]
loud_tracks
```

```python
sns.barplot(x='loudness',y='song_title',data=loud_tracks)
plt.title('top 5 loudest tracks')
plt.show()
```

Artist with the most danceability song

```python
artist_most_danceability = df.sort_values(by='danceability',ascending = False)[['danceability','artist','song_title']].iloc[:5]
artist_most_danceability
```

```python
df['danceability'].max()
```

```python
sns.barplot(x='danceability',y='artist',data=artist_most_danceability)
plt.title("artists most danceable")
plt.show()
```

Top 10 instrumentalness tracks

```python
top_instrumental_tracks = df.sort_values(by='instrumentalness', ascending = False)[['instrumentalness','artist','song_title']].iloc[:10]
top_instrumental_tracks                            
```

```python
sns.barplot(x='instrumentalness',y='song_title', data = top_instrumental_tracks)
plt.title('top_instrumental_tracks')
plt.show()
```

Multiple feature plots

```python
interest_feature_cols = ['tempo','loudness','acousticness','danceability','duration_ms','energy',
                         'instrumentalness','liveness','speechiness','valence']
```

```python
for feat_col in interest_feature_cols:
    pos_data = df[df['target']==1][feat_col]
    neg_data = df[df['target']==0][feat_col]
    
    plt.figure(figsize=(12,7))
    
    sns.distplot(pos_data, bins=30, label='Positive', color = 'green')
    sns.distplot(neg_data, bins=30, label='Negative', color = 'red')
    
    plt.legend(loc='upper right')
    plt.title(f'Positive and Negative Histogram plot for {feat_col}')
    plt.show()
```
