```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

boxcox transform helps to transform non-normal distribution to a normal distribution. But not always will the
output be normal distribution, so we need to confirm it via qq plot

https://www.geeksforgeeks.org/box-cox-transformation-using-python/

```python
data_points  = np.random.exponential(size = 1000)
```

```python
fitted_data,lam = stats.boxcox(data_points)
```

```python
sm.qqplot(fitted_data, fit = True, dist = stats.norm, line='45')
plt.show()
```

```python
sns.distplot(fitted_data, kde=True, hist=False)
```

```python
sns.kdeplot(data_points) #to get pdf
plt.show()
```

```python
sns.distplot(data_points, kde=True, hist= False)
plt.show()
```
