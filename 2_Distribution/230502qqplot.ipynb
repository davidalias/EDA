```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data_points = np.random.normal(loc = 20, scale = 5, size=300)
```

https://www.geeksforgeeks.org/qqplot-quantile-quantile-plot-in-python/

https://docs.scipy.org/doc/scipy/reference/stats.html

```python
sm.qqplot(data_points, line='45', fit = True, dist = stats.norm)
plt.show()
#fit=true implies standard fit to 'dist'
#if false we need to manually give mean n variance
```

```python
std_normal = np.random.normal(loc=0, scale= 1, size=300)
sm.qqplot(std_normal, line='45', fit = True, dist= stats.norm)
plt.show()
```
